```python
import os
import pandas as pd
```

```python
data = {
    'image_name': [],
    'label': [],
    'path': [],
}

for folder in os.listdir('./imagenet_dataset/'):
    for image in os.listdir('./imagenet_dataset/' + folder):
        data['image_name'].append(image)
        data['label'].append(folder)
        data['path'].append('images/' + folder + "/" + image)

df = pd.DataFrame(data)
df.head()
```

```python
print('{', end = " ")
i = 0
for folder in os.listdir('./images/'):
    if folder.endswith('csv'): continue
    print("\"" + folder + "\"", end = " : ")
    print(i, end = "")
    print(", ", end = "")
    i += 1
print("}")
```

```python
df.value_counts(['label'])

train = df.groupby('label').sample(frac = 0.7 ).index
df['train'] = False
df.loc[train, 'train'] = True
df
```

```python
traindf, testdf = df.groupby('train')
```
